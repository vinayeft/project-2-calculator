```python
import tkinter as tk
import math

# Function to evaluate expression
def click_button(item):
    global expression
    expression = expression + str(item)
    input_text.set(expression)

def clear_button():
    global expression
    expression = ""
    input_text.set("")

def equal_button():
    try:
        global expression
        result = str(eval(expression))
        input_text.set(result)
        expression = result
    except:
        input_text.set("Error")
        expression = ""

def scientific_function(func):
    try:
        global expression
        value = str(eval(expression))
        result = ""
        if func == "sqrt":
            result = str(math.sqrt(float(value)))
        elif func == "log":
            result = str(math.log10(float(value)))
        elif func == "ln":
            result = str(math.log(float(value)))
        elif func == "sin":
            result = str(math.sin(math.radians(float(value))))
        elif func == "cos":
            result = str(math.cos(math.radians(float(value))))
        elif func == "tan":
            result = str(math.tan(math.radians(float(value))))
        elif func == "pi":
            result = str(math.pi)
        elif func == "e":
            result = str(math.e)

        input_text.set(result)
        expression = result
    except:
        input_text.set("Error")
        expression = ""

# Main window
win = tk.Tk()
win.title("Scientific Calculator")
win.geometry("400x400")
win.resizable(False, False)

expression = ""
input_text = tk.StringVar()

# Input field
input_frame = tk.Frame(win)
input_frame.pack()

input_field = tk.Entry(input_frame, textvariable=input_text, font=('arial', 20, 'bold'), width=25, bd=5, insertwidth=4, bg="lightgrey", justify="right")
input_field.grid(row=0, column=0)
input_field.pack(ipady=10)

# Buttons frame
btns_frame = tk.Frame(win)
btns_frame.pack()

# First row
tk.Button(btns_frame, text="C", fg="black", bg="white", command=clear_button, height=2, width=10).grid(row=0, column=0)
tk.Button(btns_frame, text="(", command=lambda: click_button("("), height=2, width=10).grid(row=0, column=1)
tk.Button(btns_frame, text=")", command=lambda: click_button(")"), height=2, width=10).grid(row=0, column=2)
tk.Button(btns_frame, text="/", command=lambda: click_button("/"), height=2, width=10).grid(row=0, column=3)

# Second row
tk.Button(btns_frame, text="7", command=lambda: click_button(7), height=2, width=10).grid(row=1, column=0)
tk.Button(btns_frame, text="8", command=lambda: click_button(8), height=2, width=10).grid(row=1, column=1)
tk.Button(btns_frame, text="9", command=lambda: click_button(9), height=2, width=10).grid(row=1, column=2)
tk.Button(btns_frame, text="*", command=lambda: click_button("*"), height=2, width=10).grid(row=1, column=3)

# Third row
tk.Button(btns_frame, text="4", command=lambda: click_button(4), height=2, width=10).grid(row=2, column=0)
tk.Button(btns_frame, text="5", command=lambda: click_button(5), height=2, width=10).grid(row=2, column=1)
tk.Button(btns_frame, text="6", command=lambda: click_button(6), height=2, width=10).grid(row=2, column=2)
tk.Button(btns_frame, text="-", command=lambda: click_button("-"), height=2, width=10).grid(row=2, column=3)

# Fourth row
tk.Button(btns_frame, text="1", command=lambda: click_button(1), height=2, width=10).grid(row=3, column=0)
tk.Button(btns_frame, text="2", command=lambda: click_button(2), height=2, width=10).grid(row=3, column=1)
tk.Button(btns_frame, text="3", command=lambda: click_button(3), height=2, width=10).grid(row=3, column=2)
tk.Button(btns_frame, text="+", command=lambda: click_button("+"), height=2, width=10).grid(row=3, column=3)

# Fifth row
tk.Button(btns_frame, text="0", command=lambda: click_button(0), height=2, width=10).grid(row=4, column=0)
tk.Button(btns_frame, text=".", command=lambda: click_button("."), height=2, width=10).grid(row=4, column=1)
tk.Button(btns_frame, text="=", fg="white", bg="green", command=equal_button, height=2, width=22).grid(row=4, column=2, columnspan=2)

# Sixth row - scientific functions
tk.Button(btns_frame, text="sin", command=lambda: scientific_function("sin"), height=2, width=10).grid(row=5, column=0)
tk.Button(btns_frame, text="cos", command=lambda: scientific_function("cos"), height=2, width=10).grid(row=5, column=1)
tk.Button(btns_frame, text="tan", command=lambda: scientific_function("tan"), height=2, width=10).grid(row=5, column=2)
tk.Button(btns_frame, text="sqrt", command=lambda: scientific_function("sqrt"), height=2, width=10).grid(row=5, column=3)

# Seventh row
tk.Button(btns_frame, text="log", command=lambda: scientific_function("log"), height=2, width=10).grid(row=6, column=0)
tk.Button(btns_frame, text="ln", command=lambda: scientific_function("ln"), height=2, width=10).grid(row=6, column=1)
tk.Button(btns_frame, text="Ï€", command=lambda: scientific_function("pi"), height=2, width=10).grid(row=6, column=2)
tk.Button(btns_frame, text="e", command=lambda: scientific_function("e"), height=2, width=10).grid(row=6, column=3)

win.mainloop()
```
